```python
def search_occrp(query, doc_type, country=None):
    
    search = 'https://data.occrp.org/api/2/search?'
    url_search = search + 'q=' + query + '&filter:schemata=' + doc_type + 'filter:countries=' + country
    
    r = requests.get(url_search, headers=header).json()
    
    for r in r_search['results']:
        if r['schema'] == doc_type:
            schema = r.get('schema')
            schemata = r.get('schemata')
            score = r.get('score')
            country = r['properties'].get('country')
            website = r['properties'].get('website')
            address = r['properties'].get('address')
            name = r['properties'].get('name')
            alias = r['properties'].get('alias')
            company_id = r.get('id')
            collection_id = r['collection'].get('collection_id')
            collection_label = r['collection'].get('label')
            collection_summary = r['collection'].get('summary')
            collection_publisher = r['collection'].get('publisher')
            link_occrp = r['links'].get('ui')
        if not None:
            return schema, schemata, score, country, address, name, alias, 
              company_id, collection_id, collection_label, collection_summary, 
              collection_publisher, link_occrp
```
